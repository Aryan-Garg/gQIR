```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
REAL_IMGS_OUT = "/media/agarg54/Extreme SSD/code/gQVR/evaluation_dieDrop"
```

```python
def apply_awb_manual(img_path):
    img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB).astype(np.float32) / 1.37
    avg_b, avg_g, avg_r = np.mean(img, axis=(0,1))
    avg_gray = (avg_b + avg_g + avg_r) / 3
    scale = avg_gray / np.array([avg_b, avg_g, avg_r])
    balanced = np.clip(img * scale, 0, 255).astype(np.uint8)
    return img.astype(np.uint8), balanced

def wb_maxrgb(img):
    imgf = img.astype(np.float32)
    maxes = imgf.reshape(-1,3).max(axis=0)
    scale = 255.0 / (maxes + 1e-6)
    out = np.clip(imgf * scale, 0, 255).astype(np.uint8)
    return out


def wb_lab_neutral(img):
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB).astype(np.float32)
    L, a, b = cv2.split(lab)
    a -= a.mean()
    b -= b.mean()
    lab = cv2.merge([L, a, b])
    out = cv2.cvtColor(np.clip(lab, 0, 255).astype(np.uint8), cv2.COLOR_LAB2BGR)
    return out
```

```python
for img_name in os.listdir(REAL_IMGS_OUT):
    # print(img_name)
    if img_name.endswith(".png"):
        if "out" in img_name:
            img_path = os.path.join(REAL_IMGS_OUT, img_name)
            img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB).astype(np.float32) / 255.
            image, balanced_image = apply_awb_manual(img_path)
            save_name = 'corrected_' + img_name
            # print(f"Saved to {os.path.join(REAL_IMGS_OUT, save_name)}")
            cv2.imwrite(f"{os.path.join(REAL_IMGS_OUT, save_name)}", cv2.cvtColor(balanced_image, cv2.COLOR_RGB2BGR))
            
            # balanced_image = wb_maxrgb(img)
            # print(np.max(img), np.min(img))
            # plt.subplot(1,2,1)
            # plt.imshow(img)
            # plt.axis('off')
            # plt.title("input")

            # plt.subplot(1,2,2)
            # plt.imshow(balanced_image)
            # plt.axis('off')
            # plt.title('balanced')
            
```
